```python
#importando bibliotecas 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#importando o dataset
df = pd.read_csv('./Sales_Data.csv')
df
```

```python
df.info()
#Observa-se que na nas colunas ReturnReason, CrossSellProducts e UpsellProducts existem valores NaN. 
#Os valores com NaN dessas 3 colunas podem ser modificados por 'No', pois s√£o valores admitem essa mudan√ßa,
#Uma vez que ReturnReason √© a raz√£o de devolu√ß√£o e o valor 'No' significaria que n√£o houve devolu√ß√£o. 
#J√° as colunas CrossSellProducts e UpsellProducts avaliam se houveram ou n√£o compras conjuntas ou upgrade de equipamento
#logo, as que s√£o NaN ser√£o substituidas por 'No' para melhor an√°lise.
```

```python
#Convertendo valores 'Nan' para 'No'
df.fillna("No", inplace=True)
df.info()
```

## Analisando o total de vendas por per√≠odo (dia, m√™s, ano)
- 1Ô∏è‚É£ Converter a coluna TransactionDate para o formato de data
- 2Ô∏è‚É£ Criar colunas separadas para o Ano, M√™s e Dia
- 3Ô∏è‚É£ Agrupar os dados e calcular o total de vendas (FinalSalePrice)
- 4Ô∏è‚É£ Visualizar os resultados com gr√°ficos

```python
# Converter a coluna de data para formato datetime
df["TransactionDate"] = pd.to_datetime(df["TransactionDate"])

# Criar colunas separadas para Ano, M√™s e Dia
df["Year"] = df["TransactionDate"].dt.year
df["Month"] = df["TransactionDate"].dt.month
df["Day"] = df["TransactionDate"].dt.day


# Encontrar a data mais antiga e a mais recente
data_mais_antiga = df["TransactionDate"].min()
data_mais_recente = df["TransactionDate"].max()

print(f"üìÖ Data mais antiga: {data_mais_antiga}")
print(f"üìÖ Data mais recente: {data_mais_recente}")
```

### Com essa an√°lise, percebemos que n√£o podemos fazer muitas compara√ß√µes entre o ano de 2024 e 2025 pois foram gravados 5 meses de dados do ano de 2024 e apenas 1 m√™s do ano de 2025, n√£o sendo possivel fazer a compara√ß√£o de vendas por ano, m√™s e dia.

##  Compara√ß√£o de vendas por categoria e produto

- 1Ô∏è‚É£ Agrupar e somar o valor das vendas (FinalSalePrice) por categoria
- 2Ô∏è‚É£ Agrupar e somar o valor das vendas por produto
- 3Ô∏è‚É£ Criar gr√°ficos para visualizar os resultados

```python
# Agrupar vendas por categoria e somar as vendas
vendas_por_categoria = df.groupby("ProductCategory")["FinalSalePrice"].sum().sort_values(ascending=False)

# Agrupar vendas por produto e somar as vendas
vendas_por_produto = df.groupby("ProductName")["FinalSalePrice"].sum().sort_values(ascending=False)

# Agrupar e somar a quantidade de produtos vendidos por categoria
quantidade_por_categoria = df.groupby("ProductCategory")["SalesQuantity"].sum().sort_values(ascending=False)


# Criar os gr√°ficos
sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(3, 1, figsize=(12, 12))

# üìä Gr√°fico de vendas por categoria
sns.barplot(x=vendas_por_categoria.values, y=vendas_por_categoria.index, ax=axes[0], palette="Blues")
axes[0].set_title("Total de Vendas por Categoria")
axes[0].set_xlabel("Valor Total (R$)")
axes[0].set_ylabel("Categoria")

# üìä Gr√°fico de vendas por produto (top 10 mais vendidos)
top_10_produtos = vendas_por_produto.head(10)
sns.barplot(x=top_10_produtos.values, y=top_10_produtos.index, ax=axes[1], palette="Greens")
axes[1].set_title("Top 10 Produtos Mais Vendidos")
axes[1].set_xlabel("Valor Total (R$)")
axes[1].set_ylabel("Produto")

# Criar gr√°fico de vendas por categoria
sns.barplot(x=quantidade_por_categoria.values, y=quantidade_por_categoria.index, ax=axes[2], palette="Reds")
axes[2].set_title("Quantidade de Produtos Vendidos por Categoria")
axes[2].set_xlabel("Quantidade de Produtos Vendidos")
axes[2].set_ylabel("Categoria")

# Ajustar layout
plt.tight_layout()
plt.show()

# Calcular o Ticket M√©dio
ticket_medio = df["FinalSalePrice"].sum() / df["TransactionID"].nunique()

print(f"üí∞ Ticket M√©dio das Vendas: R$ {ticket_medio:.2f}")
```
